# Exploratory Data Analysis

```python
# import libaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# 1.1 Import Data

```python
train = pd.read_csv("train.csv")
val = pd.read_csv("val.csv")

explore = pd.concat([train, val], axis=0)
explore.head()
```

# 2.1 Univariate Analysis: Difference In Frequency When Split By Class

```python
poison = explore[explore["class"]=="p"]
edible = explore[explore["class"]=="e"]

for col in explore.columns:
    if col != "class":
        
        print(f"\n\n\n{col}\n")
        
        poison_bar_data = sorted(dict(poison[col].value_counts(normalize=True)).items())
        edible_bar_data = sorted(dict(edible[col].value_counts(normalize=True)).items())
        
        poison_labels = []
        poison_values = []
        for key, value in poison_bar_data:
            poison_labels.append(key)
            poison_values.append(value)
            
        edible_labels = []
        edible_values = []
        for key, value in edible_bar_data:
            edible_labels.append(key)
            edible_values.append(value)        
        
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 8))
        ax1.bar(poison_labels, poison_values)
        ax1.set_title("Poison")
        ax2.bar(edible_labels, edible_values)
        ax2.set_title("Edible")
        
        plt.show()
        
        
```

It appears all the features give insight into classification, except veil type.
